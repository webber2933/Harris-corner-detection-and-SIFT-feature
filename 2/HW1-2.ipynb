# SIFT interest point detection and matching

A. SIFT interest point detection

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# function of img show
def imshow(image):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    else:
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
img1 = cv2.imread('1a_notredame.jpg')
img2 = cv2.imread('1b_notredame.jpg')
```

```python
sift = cv2.xfeatures2d.SIFT_create(100)
keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
keypoints2, descriptors2 = sift.detectAndCompute(img2, None)
print(descriptors1)
```

```python
img1_interest_points = cv2.imread('1a_notredame.jpg')
img2_interest_points = cv2.imread('1b_notredame.jpg')
cv2.drawKeypoints(img1_interest_points,keypoints1,img1_interest_points,(0,255,255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.drawKeypoints(img2_interest_points,keypoints2,img2_interest_points,(0,255,255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
imshow(img1_interest_points)
imshow(img2_interest_points)
cv2.imwrite('./output/img1_interest_points.jpg',img1_interest_points)
cv2.imwrite('./output/img2_interest_points.jpg',img2_interest_points)
```

B. SIFT feature matching

```python
# find the Euclidean distances between 2 descriptors
def Euclidean_dist(des1,des2):
    sum_sq = np.sum(np.square(des1 - des2))
    return np.sqrt(sum_sq)
```

```python
# finds a list of interest point correspondences based on nearest-neighbor matching principle

def nearest_neighbor_matching(descriptors1,descriptors2):

    matches = []

    for i in range(len(descriptors1)):
        all_dist = []
        ref_descriptor = descriptors1[i]
        for j in range(len(descriptors2)):
            dist = Euclidean_dist(ref_descriptor,descriptors2[j])
            all_dist.append([j,dist])

        all_dist = sorted(all_dist,key = lambda x: x[1])

        matches.append([i,all_dist[0][0]])
            
    return matches
```

```python
def draw_matching(matches,img1,img2):
    # combine 2 images
    black = np.zeros((img1.shape[0] - img2.shape[0], img2.shape[1],3), dtype=np.uint8)
    img2_padding = np.concatenate((img2,black))
    combine_img = np.concatenate((img1,img2_padding), axis=1)

    # draw matching lines
    for line in matches:
        pt1 = keypoints1[line[0]].pt
        int_pt1 = (int(pt1[0]),int(pt1[1]))
        pt2 = keypoints2[line[1]].pt
        int_pt2 = (int(pt2[0]) + img1.shape[1],int(pt2[1]))

        cv2.line(combine_img, int_pt1, int_pt2, (0, 0, 255), 3)

    imshow(combine_img)
    return combine_img
```

```python
# nearest_neighbor_matching
matches = nearest_neighbor_matching(descriptors1,descriptors2)
feature_matching = draw_matching(matches,img1,img2)
cv2.imwrite('./output/feature_matching.jpg',feature_matching)
print(len(matches))
```

reduce the mis-matching

```python
# finds a list of interest point correspondences based on nearest-neighbor matching principle with ratio-test

def nearest_neighbor_matching_ratio_test(descriptors1,descriptors2,threshold):

    matches = []
    best_matches = []

    for i in range(len(descriptors1)):
        all_dist = []
        ref_descriptor = descriptors1[i]
        for j in range(len(descriptors2)):
            dist = Euclidean_dist(ref_descriptor,descriptors2[j])
            all_dist.append([j,dist])

        all_dist = sorted(all_dist,key = lambda x: x[1])
        
        min_idx = all_dist[0][0]
        min_dist = all_dist[0][1]
        second_dist = all_dist[1][1]
        
        matches.append([i,min_idx,min_dist,second_dist])

        #if all_dist[0][1] / all_dist[1][1] < threshold:
        #    matches.append([i,all_dist[0][0]])
        
    # take the top 25% matches
    matches = sorted(matches,key = lambda x: x[2])
    wanted_num = int(len(matches) * 0.25)
    matches = matches[:wanted_num]
    print(len(matches))
    
    # ratio test
    for match in matches:
        if match[2] / match[3] < threshold:
            best_matches.append([match[0],match[1]])
            
    return best_matches
```

```python
# find more keypoints
sift = cv2.xfeatures2d.SIFT_create(2000)
keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
keypoints2, descriptors2 = sift.detectAndCompute(img2, None)

matches = nearest_neighbor_matching_ratio_test(descriptors1,descriptors2,0.7)
feature_matching_reduce_mismatching = draw_matching(matches,img1,img2)
cv2.imwrite('./output/feature_matching_reduce_mismatching.jpg',feature_matching_reduce_mismatching)
print(len(matches))
```
